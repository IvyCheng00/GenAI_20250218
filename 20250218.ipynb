```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def draw_spirograph(R=5, r=3, d=5, num_points=1000):
    # 時間參數
    t = np.linspace(0, 10 * np.pi, num_points)
    # 函數
    x = (R - r) * np.cos(t) + d * np.cos((R - r) / r * t)
    y = (R - r) * np.sin(t) - d * np.sin((R - r) / r * t)

    # 建立 DataFrame，進行數據分析
    df = pd.DataFrame({'x': x, 'y': y})

    # 設定背景風格
    sns.set(style="whitegrid")

    # 繪製圖形
    plt.figure(figsize=(6, 6))
    plt.plot(df['x'], df['y'], 'm', linewidth=2)

    # 加入網格線
    plt.grid(True, linestyle='--', alpha=0.6)

    # 設定圖形比例、標題
    plt.axis("equal")
    plt.title("Hypotrochoid", fontsize=14, fontweight='bold', color='m')

    plt.show()

    return df

# 執行函數並儲存 DataFrame
df_spiro = draw_spirograph()
```
